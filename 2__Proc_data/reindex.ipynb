```python
import numpy as np
import pandas as pd

import os
import sys
```

```python
PATH_TO_SEISM_SIGNAL = "../1__Preproc_data/Seism_analise/Wells_seism.csv" 
PATH_TO_SEISM_Z = "../1__Preproc_data/Seism_analise/Wells_index_seism.csv" 
df_sig = pd.read_csv(PATH_TO_SEISM_SIGNAL, sep=";")
df_sig.set_index("z_m_index", inplace=True)

df_z = pd.read_csv(PATH_TO_SEISM_Z, sep=";")
df_z.set_index("z_m_index", inplace=True)
```

```python
df_sig
```

```python
df_z
```

```python
z_m_new = np.arange(0, 3729, 4)
new_df_matrix = []
for col in df_z.columns:
    new_df_matrix.append(np.interp(z_m_new, df_z[col], df_sig[col]))

new_df_matrix = np.array(new_df_matrix).T
```

```python
df_reindex_seism = pd.DataFrame(new_df_matrix, index = z_m_new, columns=df_z.columns)
df_reindex_seism
```

```python
df_reindex_seism.to_csv("Reindex_seism.csv", sep=";")
df_reindex_seism.to_csv("Reindex_seism.xlsx")
```
